#### val의 song을 기반으로 tag 붙이기 (song에서 tag를 만듬)

```python
val = pd.read_json('datas/val.json', typ = 'frame')
val
```

```python
# 빈 리스트 확인
empty = []
for j in val.columns:
    for i in range(0, len(val)):
        if not val[j][i]:
            empty.append({j: i})
            print(f'{j} 의 {i}번째가 비어있습니다.')
```

```python
# [df_idx][song_idx]
val['songs'][0][0]
```

```python
from collections import Counter
from tqdm import notebook
val_copy = val.copy()
# df_indx = i
for i in notebook.tqdm(range(0,len(val_copy))):
    tags = {}
    if len(val_copy['songs'][i]) != 0:
        # song_idx = j
        for j in range(0,len(val_copy['songs'][i])):
            if len(song_tags[song_tags['songs'] == val_copy['songs'][i][j]]) != 0:
                tag = song_tags[song_tags['songs'] == val_copy['songs'][i][j]]['tags_num'].values[0]
                tags = Counter(tags) + Counter(tag)
            val_copy['tags'][i] = list(dict(sorted(tags.items(), key = lambda x:x[1],reverse=True)[:10]).keys())
    else :
        print(f'{i}번째 노래없음')
```

```python
from collections import Counter
from tqdm import notebook
val_copy2 = val_copy.copy()
# df_indx = i
for i in notebook.tqdm(range(0,len(val_copy2))):
    songs = {}
    if len(val_copy2['tags'][i]) != 0:
        # song_idx = j
        for j in range(0,len(val_copy2['tags'][i])):
            if len(tag_songs[tag_songs['tags'] == val_copy2['tags'][i][j]]) != 0:
                song = tag_songs[tag_songs['tags'] == val_copy2['tags'][i][j]]['songs_num'].values[0]
                songs = Counter(songs) + Counter(song)
                val_copy2['songs'][i] = list(dict(sorted(songs.items(), key = lambda x:x[1],reverse=True)[:100]).keys())
    else :
        print(f'{i}번째 태그없음')
```

```python
from collections import Counter
from tqdm import notebook
val_copy3 = val_copy2.copy()
# df_indx = i
for i in notebook.tqdm(range(0,len(val_copy3))):
    if len(val_copy3['tags'][i]) != 10 and len(val_copy3['tags'][i]) !=0:
        if len(val_copy3['songs'][i]) != 0:
            tags = {}
            # song_idx = j
            for j in range(0,len(val_copy3['songs'][i])):
                if len(song_tags[song_tags['songs'] == int(val_copy3['songs'][i][j])]) != 0:
                    tag = song_tags[song_tags['songs'] == int(val_copy3['songs'][i][j])]['tags_num'].values[0]
                    tags = Counter(tags) + Counter(tag)
                    val_copy3['tags'][i] = list(dict(sorted(tags.items(), key = lambda x:x[1],reverse=True)[:10]).keys())
```

```python
# 확인
empty = []
for i in range(0,len(val_copy3)):
    if len(val_copy3['tags'][i]) != 10 and len(val_copy3['tags'][i]) !=0:
        empty.append(i)
        print(f'{i}번째 태그가 1개 ~ 9개입니다.')
```

```python
# 확인
# 태그가 1 ~ 9개 사이이지만, 음악이 채워져있다면 위의 코드가 잘못된것이기 떄문에 확인할 필요가있음
# 다만 태그가 1 ~ 9개 사이면서. 음악이 없다면 이는 음악이 없어서 태그를 못채우고, 태그는 맞는게 없어서 음악을 못채우는것
empty_song = []
for i in empty:
    if len(val_copy3['songs'][i]) != 0:
        empty_song.append(i)
        print(f'{i}번째 음악이 비어있지않습니다.')
```
