## Validation, Test의 태그 또는 노래가 없는 플레이리스트 찾기
- 태그가 없는 플레이리스트 태그 추가
-  노래가 없는 플레이리스트 노래 추가

### 1. 태그가 없는 플레이리스트 태그 추가
- 1. 플레이리스트에 노래가 있다면 노래들에서 가장 많이 언급된 장르를 찾고 또 그 장르에서 가장 많이 언급된 태그를 추가. (제목으로부터 태그를 유추하는 것보다 정확함)

- 2. 노래가 없다면 플레이리스트 제목에서 명사와 형태소를 추출하여 그 명사와 형태소의 이름을 가진 태그가 있다면 그 태그를 모두 추가. (명사와 형태소를 두개 모두 사용한 이유는 명사만 추출하게되면 형용사, 부사 같은 경우는 추출하지 못하기 때문에 형태소도 같이 추출)

- 3. 그럼에도 태그가 1개라도 채워지지 않은 플레이리스트가 있다면 모든 태그를 TFIDF로 벡터화 시킨 후 플레이리스트 제목과의 거리를 비교해서 가장 가까운 태그 1개 추가

- 4. 그럼에도 태그가 1개라도 채워지지 않은 플레이리스트는 제목이 없는 것이기 때문에 유추할 수 없다고 판단하여 전체플레이리스트에서 가장 많이 언급된 태그(기분전환)를 추가.

#### 세부장르별 언급된 태그의 갯수를 세어 가장 많이 언급된 태그 1개만 추가


#### 플레이리스트 노래들의 가장 많이 언급된 장르를 새로운 컬럼에 추가해주고 그 장르에서 가장 많이 언급된 태그를 추가


#### 노래가 존재하지 않는 플레이리스트


#### 명사와 형태소를 추가하고 그 명사와 형태소의 이름을 가진 태그가 존재한다면 모두 추가
- 명사만 뽑으면 형용사, 부사 등 중요하게 생각되는 키워드를 놓칠 수 있기에 모두 추출


#### 그럼에도 불구하고 태그를 1개라도 채우지 못한 플레이리스트


#### 모든 태그를 TFIDF로 벡터화 시킨 후 플레이리스트 제목과의 거리를 구해 가장 가까운 거리의 태그를 1개 추가


#### 그럼에도 태그가 1개도 안채워진 플레이리스트는 전체 플레이리스트에서 가장 많이 언급된 태그(기분전환)을 추가


### 2. 노래가 없는 플레이리스트 노래 추가
- 노래가 없는 플레이리스트의 노래들은 플레이리스트 제목으로 노래를 유추하게 된다면 정확성이 떨어져 처음에 시도했던 태그별 노래가 많이 언급된 상위 10개의 노래들을 추가
