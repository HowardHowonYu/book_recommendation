```python
og_raw_data = pd.read_csv('./bookcontents4.csv', error_bad_lines=False)
```

```python
og_raw_data.columns
```

```python
df = og_raw_data.dropna(axis=0)
```

```python
df[['id','text']]
```

```python
from krwordrank.hangle import normalize

# texts = ['이것은 예문입니다', '각자의 데이터를 준비하세요', ... ]
texts = df['text'].iloc[:3].values
texts = [normalize(text, english=True, number=True) for text in texts]
```

```python
from krwordrank.word import KRWordRank

wordrank_extractor = KRWordRank(
    min_count = 1, # 단어의 최소 출현 빈도수 (그래프 생성 시)
    max_length = 10, # 단어의 최대 길이
    verbose = True
    )

beta = 0.85    # PageRank의 decaying factor beta
max_iter = 10

keywords, rank, graph = wordrank_extractor.extract(texts, beta, max_iter)
```

```python
keywords,rank
```

```python
for word, r in sorted(keywords.items(), key=lambda x:x[1], reverse=True)[:30]:
    print('%8s:\t%.4f' % (word, r))
```

```python
stopwords = {'나는', '있는', '것이', '내가', '하지만'}
passwords = {word:score for word, score in sorted(
    keywords.items(), key=lambda x:-x[1])[:300] if not (word in stopwords)}
```
